### Correlated Subquery

```python
query1 = '''
SELECT e1.empname, e1.dept, max(e1.salary) as salary2nd FROM emp e1 where e1.salary < (
    SELECT max(e2.salary) FROM emp e2 WHERE e2.dept = e1.dept
) group by e1.dept;
'''

conn = sqlite3.connect('employee.db')
cursor = conn.cursor()
df = pd.read_sql_query(query1, conn)
conn.close()
df
```

### Windows Function

```python
query3 = '''SELECT empname, dept, salary
from (SELECT empname, dept, salary, DENSE_RANK() OVER(PARTITION BY dept ORDER BY salary DESC) AS RNK FROM emp) as e
WHERE RNK = 2;'''

conn = sqlite3.connect('employee.db')
cursor = conn.cursor()
df3 = pd.read_sql_query(query3, conn)
conn.close()
df3
```
